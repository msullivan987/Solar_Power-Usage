```python
%matplotlib notebook

import pandas as pd 
import datetime as dt
import numpy as np
import matplotlib.pyplot as plt
pd.options.mode.chained_assignment = None  # default='warn'


excel_file = "power generation average set.xls"

power_gen_df = pd.read_excel(excel_file)

power_gen = power_gen_df[["Unnamed: 1","Unnamed: 13"]]

power_g = power_gen.rename(columns={"Unnamed: 1":"State","Unnamed: 13":"Grand Total"})
power_g = power_g.drop(power_g.index[[0,1,2,3,51,52]])
power_g

```

```python
x_axis = np.arange(len(power_g))
tick_locations = [value+0.4 for value in x_axis]

plt.figure(figsize=(15,3))
plt.bar(x_axis, power_g["Grand Total"], color='r', alpha=0.5, align="edge")
plt.xticks(tick_locations, power_g["State"], rotation="vertical")

plt.title("2009-2019 Average Solar Output by State")
plt.xlabel("State")
plt.ylabel("Avg. Megawatt Hour Output")
```
